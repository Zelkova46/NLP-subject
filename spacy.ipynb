```python

import spacy
```

```python
nlp = spacy.load('en_core_web_sm')
```

```python
s = nlp("I shot an elephant in my pajamas")
  
```

```python
print(type(s))
```

```python
#POS Tagging
for token in s:
    print(token.text, token.pos_, token.tag_)
     
```

```python
#Chunking

for chunk in s.noun_chunks:
    print(chunk.text, chunk.label_, chunk.root.text)
```

```python
#Dependency Parsing

for token in s:
    print("{0}/{1} <-- {2} <-- {3}/{4}".format(
        token.text, token.tag_, token.dep_, token.head.text, token.head.tag_
    ))
     
```

```python
#Dependecy Parsing Visualization

from spacy import displacy

displacy.render(s, style='dep', jupyter=True, options={'distance':85})
     
```

```python
import csv
import xmltodict
```

```python
with open("/content/2010-blitdraft.xml", "r") as file:
  filedata = file.read()

#convverting xml to python dictionary (ordered dict)
data_dict = xmltodict.parse(filedata)
```

```python
data_dict.keys()
```

```python
data_dict = data_dict['req_document']
```

```python
data_dict.keys()
```

```python
data_dict = data_dict['p']
```

```python
print((data_dict))
print(len(data_dict))
```

```python
for item in data_dict:
  print(item.keys())
```

```python
sens = []
data = []
```

```python
for item in data_dict:
  for value in item.keys():
    if value == 'p':
      data.append(item['p'])
```

```python
data = data[0]
```

```python
data
```

```python
for item in data:
  for value in item.keys():
    if value == 'text_body':
      sens.append(item['text_body'])
```

```python
sens
```

```python
hasil = []
for sub in sens:
  hasil.append(sub.replace("\n", ""))
```

```python
temp = []
for value in hasil:
  temp.append(value.split('. '))
```

```python
sentences = []
for value in temp:
  for item in value:
    sentences.append(item)
```

```python
print(sentences)
```

```python
len(sentences)
```

```python
sentences[0:2]
```

```python
S_xml= []

for load in range(len(sentences)):
  S_xml.append(nlp(sentences[load]))
```

```python
print(type(S_xml[0]))
```

POS TAGGING

```python
for doc in S_xml:
    
    print("POS tagging perkalimat :")
    for token in doc:

        print(token.text, token.pos_, token.tag_, token.dep_)
    print("\n")
```

CHUNKING

```python

#Chunking

for x in range(len(S_xml)):
  print("Chunking perkalimat")
  for chunk in S_xml[x].noun_chunks:
      print(chunk.text, chunk.label_, chunk.root.text)

  print("\n")
```

```python

#Dependency Parsing

for x in range(len(S_xml)):
  print("Dependency parsing pekalimat")
  for token in S_xml[x]:
      print("{0}/{1} <-- {2} <-- {3}/{4}".format(
          token.text, token.tag_, token.dep_, token.head.text, token.head.tag_
      ))
  print("\n")
```

```python

#Dependecy Parsing Visualization

from spacy import displacy

for x in range(len(S_xml)):
  print("Parsing tiap kalimat \n")
  displacy.render(S_xml[x], style='dep', jupyter=True, options={'distance':85})
```
